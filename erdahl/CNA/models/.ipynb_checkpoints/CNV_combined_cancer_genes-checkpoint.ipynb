This file contains analysis of CNV data as well as integrated with WPR data from Ted.

Several sections for generating desired output are at the end for Erdahl.

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
import shap
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import SpectralBiclustering
import numpy as np
```

```python
labels = {0:'W', 1:'M', 2:'T'}
```

# CNV only

## Random Forest

```python
C = pd.read_csv('./C.csv').set_index(['sample'])
cancer = pd.read_csv('../data/ref_data/CancerDriverGenes.tsv', sep = '\t', header=None, names=['cancer_genes'])
```

```python
X = C.drop(['y'], axis=1)
X = X[X.columns.intersection(cancer['cancer_genes'].values)]
y = C['y']
```

```python
rf = RandomForestClassifier(n_estimators=5000,oob_score=True, n_jobs=-1)
rf.fit(X, y)
feature_importances = pd.DataFrame(
    rf.feature_importances_, index=X.columns,
    columns=['importance']).sort_values(
        'importance', ascending=False)
rf.oob_score_
```

```python
feature_importances.to_csv('./simon_results/C_RF.csv',index_label='Gene')
```

```python
by_class = X[X.columns.intersection(list(feature_importances.iloc[:30,0].index))]
by_class['class'] = y
by_class = by_class.sort_values(by=['class'])
by_class['class'] = by_class['class'].map(lambda x:labels[x])
by_class = by_class.reset_index()
by_class['sample'] = by_class.apply(lambda row:row['sample']+'_'+row['class'], axis=1)
by_class = by_class.drop(['class'], axis=1)
by_class = by_class.set_index('sample')
fig = plt.gcf()
fig.set_size_inches(15, 15)
ax = sns.heatmap(by_class, annot=True, cmap='RdBu_r', vmax=2, vmin= -2)
plt.savefig('./simon_results/by_class/C_RF.png')
```

```python
rf_explaniner = shap.TreeExplainer(rf)
shap_values = rf_explaniner.shap_values(X)
plt.title('C_RF')
shap.summary_plot(shap_values, X,max_display=30, show=False)
plt.savefig('./simon_results/shaps/C_RF.png')
```

## XGBoost

```python
xgb = XGBClassifier(n_estimators=120, n_jobs=-1)
xgb.fit(X, y)
feature_importances = pd.DataFrame(
    xgb.feature_importances_, index=X.columns,
    columns=['importance']).sort_values(
        'importance', ascending=False)
feature_importances.to_csv('./simon_results/C_XGB.csv',index_label='Gene')
```

```python
by_class = X[X.columns.intersection(list(feature_importances.iloc[:30,0].index))]
by_class['class'] = y
by_class = by_class.sort_values(by=['class'])
by_class['class'] = by_class['class'].map(lambda x:labels[x])
by_class = by_class.reset_index()
by_class['sample'] = by_class.apply(lambda row:row['sample']+'_'+row['class'], axis=1)
by_class = by_class.drop(['class'], axis=1)
by_class = by_class.set_index('sample')
fig = plt.gcf()
fig.set_size_inches(15, 15)
ax = sns.heatmap(by_class, annot=True, cmap='RdBu_r', vmax=2, vmin= -2)
plt.savefig('./simon_results/by_class/C_XGB.png')
```

```python
xgb_explaniner = shap.TreeExplainer(xgb)
shap_values = xgb_explaniner.shap_values(X)
plt.title('C_XGB')
shap.summary_plot(shap_values, X,max_display=30, show=False)
plt.savefig('./simon_results/shaps/C_XGB.png')
```

# WPRC subset

```python
SV40 = pd.read_csv('../data/ref_data/SV40_HPV_vector_integration_table.csv')
```

```python
SV40['Breakpoint_gene'] = SV40['Breakpoint_gene'].str.strip()
SV40_genes = SV40['Breakpoint_gene'].unique().tolist()
```

```python
SV40_genes_W = ['W_'+str(x) for x in SV40_genes]
SV40_genes_P = ['P_'+str(x) for x in SV40_genes]
SV40_genes_R = ['R_'+str(x) for x in SV40_genes]
SV40_genes_C = ['C_'+str(x) for x in SV40_genes]
SV40_genes_all = SV40_genes_W + SV40_genes_P + SV40_genes_R + SV40_genes_C
```

```python
W = pd.read_csv('./W.tsv', sep = '\t')
P = pd.read_csv('./P.tsv', sep = '\t')
R = pd.read_csv('./R.tsv', sep = '\t')
C = pd.read_csv('./C.csv').sort_values(['sample']).set_index(['sample'])
cancer = pd.read_csv('../data/ref_data/CancerDriverGenes.tsv', sep = '\t', header=None, names=['cancer_genes'])
```

```python
C.columns =  ['C_' + str(col) for col in C.columns if str(col) != 'y'] + ['y']
```

```python
X = W.drop(['West'], axis=1)
X = pd.concat([X, R], axis=1)
X = pd.concat([X, P], axis=1)
X.index.names = ['sample']
X = pd.merge(X, C, on=['sample'])
y = X['y']
X = X.drop(['y'],axis=1)
```

```python
by_class = X[X.columns.intersection(SV40_genes_all)]
by_class['class'] = y
by_class = by_class.sort_values(by=['class'])
by_class['class'] = by_class['class'].map(lambda x:labels[x])
by_class = by_class.reset_index()
by_class['sample'] = by_class.apply(lambda row:row['sample']+'_'+row['class'], axis=1)
by_class = by_class.drop(['class'], axis=1)
by_class = by_class.set_index('sample')
# by_class.to_excel('./simon_results/excel/WRPC_without_ATRX_RF_100.xlsx', 'Top100')
```

```python
col_d = dict(zip(range(len(by_class.columns)), by_class.columns))
```

```python
clustering = SpectralBiclustering(n_clusters=10).fit(by_class)
fit_data = by_class.loc[:, [col_d[i] for i in np.argsort(clustering.column_labels_)]]
fig = plt.gcf()
fig.set_size_inches(15, 15)
ax = sns.heatmap(fit_data, annot=True, cmap='RdBu_r', vmax=2, vmin= -2)
```

```python
clustering = SpectralBiclustering(n_clusters=3).fit(by_class)
fit_data = by_class.loc[:, [col_d[i] for i in np.argsort(clustering.column_labels_)]]
fit_data.to_excel('./simon_results/excel/SV40.xlsx', 'SV40')
```
