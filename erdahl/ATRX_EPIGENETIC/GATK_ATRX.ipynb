```python
import pandas as pd
```

```python
GATK_HIGH = pd.read_csv('./Raw_GATK_HIGH.csv', sep='\t')
GATK_MODERATE = pd.read_csv('./Raw_GATK_MODERATE.csv', sep='\t')
GATK_HIGH['Pos1'] = GATK_HIGH['Pos1'].astype(float)
GATK_MODERATE['Pos1'] = GATK_MODERATE['Pos1'].astype(float)
ALL = pd.concat([GATK_HIGH, GATK_MODERATE])
```

```python
ALL.shape
```

```python
sample_list = [
    'IIICF_E6E7_A1', 'IIICF_T_C3', 'IIICF_d2', 'IIICF_E6_A1', 'IIICF_402DE_D2',
    'IIICF_E6E7_C4_post', 'MeT_4A_post', 'LFS_05F_24_post'
]
```

```python
sample_map = {
    'IIICF_E6E7_A1':'U_IIICF_E6E7_A1',
    'IIICF_T_C3':'U_IIICF_T_C3',
    'IIICF_d2':'U_IIICF_d2',
    'IIICF_E6_A1':'U_IIICF_E6_A1',
    'IIICF_402DE_D2':'N_IIICF_402DE_D2',
    'IIICF_E6E7_C4_post':'N_IIICF_E6E7_C4_post',
    'MeT_4A_post':'N_MeT_4A_post',
    'LFS_05F_24_post':'N_LFS_05F_24_post',
}
```

```python
ALL = ALL[ALL['Name'].isin(sample_list)]
ALL['Name'] = ALL['Name'].map(lambda x:sample_map[x])
```

## TF<sub>ATRX</sub> analysis

### p=0.00005

```python
gene_list = [
    'ASCL1', 'ATF6', 'EGR1', 'ESRRA', 'FOXS1', 'GLI1', 'HAND1', 'HESX1',
    'HOXA2', 'HOXA9', 'HOXD8', 'HSFX1', 'KLF8', 'MYBL1', 'MYOD1', 'SOX7',
    'TBX19', 'TCF7L1', 'THRB', 'ZBTB6', 'ZNF213', 'ZNF280A', 'ZNF287',
    'ZNF561', 'ZNF562', 'ZNF582'
]
gene_set = pd.read_csv(
    '../../bitbucket/CNA/data/ref_data/EnsemblGRCh37Genes.tsv', sep='\t')
p5 = gene_set[gene_set['GeneName'].isin(gene_list)].reset_index(drop=True)
```

```python
df_list = []
for i in range(p5.shape[0]):
    start = float(p5.iloc[i, 3])
    end = float(p5.iloc[i, 4])
    chromosome = 'chr' + str(p5.iloc[i, 2])
    print('checking:', p5.iloc[i, 1], chromosome + ', ', start, end)
    data = ALL[ALL['Chr1'] == chromosome]
    df = data[(start <= data['Pos1']) & (data['Pos1'] <= end)]
    df['Gene_fromScript'] = p5.iloc[i, 1]
    df_list.append(df)
df_combined = pd.concat(df_list)
```

```python
df_combined = df_combined.sort_values(by=['Name']).reset_index(drop=True)
df_combined
```

```python
writer = pd.ExcelWriter('./GATK_TF.xlsx', engine = 'xlsxwriter')
df_combined.to_excel(writer, sheet_name = 'p=0.00005', index=False)
```
