```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import collections as mc
import numpy as np
from tqdm import tqdm_notebook
```

# load gene location

```python
gene = pd.read_csv("../data/EnsemblGRCh38Genes.csv")
```

# load sample mapping

```python
samples = pd.read_excel("../data/CNVSanger/Samplesheet_SampleNames.xlsx")
samples = samples[samples['Cell_Line_Master::SIDM'].notnull()]
```

```python
samples_mapping = dict(
    zip(samples['Cell_Line_Master::SIDM'].values,
        samples['Cell Line Name'].values))
```

# load data

## plot CNV

```python
df = pd.read_csv("../data/CNVSanger/copynumber_total_new_map.csv", low_memory=False, index_col=0)

df = df.rename(columns = samples_mapping)

df = pd.melt(df.reset_index(), id_vars='index')
df.columns = ['GeneName', 'Sample', 'CN']
```

```python
df_full = pd.merge(df, gene, on=['GeneName'], how='left')
```

```python
alt_samples = [
    'G-292-Clone-A141B1', 'SK-LU-1', 'HuO9', 'Saos-2', 'SK-N-FI', 'U-2-OS',
    'NOS-1', 'NY', 'P31-FUJ', 'ES5', 'EMC-BAC-1', 'Hs746T', 'CAL-72', 'RH-41',
    'CAL-78'
]
```

```python
df_alt = df_full[df_full['Sample'].isin(alt_samples)].reset_index(drop=True)
```

```python
df_alt = df_alt[~df_alt['GeneID'].isnull()]
```

```python
df_alt_7 = df_alt[df_alt['Chrom'] == '7']
```

```python
fig, ax = plt.subplots(df_alt_7['Sample'].unique().size, 1,
                       figsize=(15, 5 * df_alt_7['Sample'].unique().size))
counter = 0
for sample in sorted(df_alt_7['Sample'].unique()):
    df_current = df_alt_7[df_alt_7['Sample'] == sample].reset_index(drop=True)
    lines = []
    colors = []

    for index, row in df_current.iterrows():
        color = 'blue' if row[
            'CN'] == 2 else 'green' if row['CN'] < 2 else 'red'
        lines.append([(row['Start'], row['CN']), (row['End'], row['CN'])])
        colors.append(color)
    lc = mc.LineCollection(lines, colors=colors, linewidths=2)
    #     fig, ax = plt.subplots(figsize=(15,3))
    ax[counter].add_collection(lc)
    ax[counter].set_title(f"{sample} Chr7")
    ax[counter].autoscale()
    ax[counter].xaxis.set_visible(False)
    counter += 1

# plt.gca().set_axis_off()
# plt.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0, 
#             hspace = 0, wspace = 0)
plt.savefig('../plots/ALT_Sanger_Chr7.pdf', orientation='portrait',bbox_inches='tight')
plt.show()
```

## calculate genome ploidy

```python
ploidies = {}
```

```python
for sample in sorted(df_alt['Sample'].unique()):
    total_len = 0
    total_ploidy = 0
    df_current = df_alt[df_alt['Sample'] == sample].reset_index(drop=True)

    for index, row in df_current.iterrows():
        total_ploidy += row['CN'] * (row['End']-row['Start'])
        total_len += (row['End']-row['Start'])
    print(f"{sample} avg ploidy: {total_ploidy/total_len:.4f}")
    ploidies[sample] = round(total_ploidy/total_len)
```

## replot with ploidy norm

```python
df = pd.read_csv("../data/CNVSanger/copynumber_total_new_map.csv", low_memory=False, index_col=0)

df = df.rename(columns = samples_mapping)

df = pd.melt(df.reset_index(), id_vars='index')
df.columns = ['GeneName', 'Sample', 'CN']

df_full = pd.merge(df, gene, on=['GeneName'], how='left')

alt_samples = [
    'G-292-Clone-A141B1', 'SK-LU-1', 'HuO9', 'Saos-2', 'SK-N-FI', 'U-2-OS',
    'NOS-1', 'NY', 'P31-FUJ', 'ES5', 'EMC-BAC-1', 'Hs746T', 'CAL-72', 'RH-41',
    'CAL-78'
]

df_alt = df_full[df_full['Sample'].isin(alt_samples)].reset_index(drop=True)

df_alt_7 = df_alt[df_alt['Chrom'] == '7']
```

```python
fig, ax = plt.subplots(df_alt_7['Sample'].unique().size, 1,
                       figsize=(15, 5 * df_alt_7['Sample'].unique().size))
counter = 0
for sample in sorted(df_alt_7['Sample'].unique()):
    df_current = df_alt_7[df_alt_7['Sample'] == sample].reset_index(drop=True)
    lines = []
    colors = []

    for index, row in df_current.iterrows():
        color = 'blue' if row[
            'CN'] == ploidies[sample] else 'green' if row['CN'] < ploidies[sample] else 'red'
        lines.append([(row['Start'], row['CN']), (row['End'], row['CN'])])
        colors.append(color)
    lc = mc.LineCollection(lines, colors=colors, linewidths=2)
    #     fig, ax = plt.subplots(figsize=(15,3))
    ax[counter].add_collection(lc)
    ax[counter].set_title(f"{sample} Chr7")
    ax[counter].autoscale()
    ax[counter].xaxis.set_visible(False)
    counter += 1

plt.savefig('../plots/ALT_Sanger_Chr7_ploidy_norm_whole_genome.pdf', orientation='portrait',bbox_inches='tight')
plt.show()
```

# convert to gistic

```python
df = pd.read_csv("../data/CNVSanger/Summary_segmentation_data_994_lines_picnic.txt", sep='\t')
```

```python
df.head()
```

```python
ploidies = {}

for sample in tqdm_notebook(sorted(df['cellLine'].unique())):
    total_len = 0
    total_ploidy = 0
    df_current = df[df['cellLine'] == sample].reset_index(drop=True)

    for index, row in df_current.iterrows():
        total_ploidy += row['totalCN'] * (row['endpos']-row['startpos'])
        total_len += (row['endpos']-row['startpos'])
#     print(f"{sample} avg ploidy: {total_ploidy/total_len:.4f}")
    ploidies[sample] = round(total_ploidy/total_len)
```

```python
sanger_alt = [
    'P31-FUJ', 'CAL-78', 'SK-LU-1', 'SK-N-FI',
    'G-292 Clone A141B1', 'HuO9', 'Saos-2', 'U-2-OS', 'NOS-1', 'NY', 'CAL-72',
    'RH-41'
]
len(sanger_alt)
```

```python
df_alt = df[df['cellLine'].isin(sanger_alt)]
```

```python
df_alt['cellLine'].unique().size
```

```python
df_alt['len'] = df_alt['SNPend'] - df_alt['SNPstart'] + 1

df_alt['totalCN'] = df_alt['totalCN'].map(lambda x:0.0001 if x == 0 else x)

df_alt['logR'] = df_alt.apply(lambda row:np.log2(row['totalCN']/ploidies[row['cellLine']]), axis=1)

df_alt_gistic = df_alt[['cellLine', 'chr', 'startpos', 'endpos', 'len', 'logR']].reset_index(drop=True)

df_alt_gistic.to_csv('../gistic/gistic_segs/cnv_gistic_sanger_ALT.txt', sep='\t', index=False, header=False)
```
