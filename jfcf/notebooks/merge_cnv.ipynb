```python
import pandas as pd
from glob import glob
```

```python
df_list = []
```

```python
for f in glob("../data/CNA/data/genes_cnv/*"):
    if 'JFCF' in f:
        df_list.append(pd.read_csv(f))
```

```python
df_JFCF = pd.concat(df_list)
```
