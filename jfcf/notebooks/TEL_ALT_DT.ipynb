```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import SpectralBiclustering
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_graphviz
import graphviz
```

```python
df = pd.read_csv("../data/normalised_counts_forpy.csv")
```

```python
X = df.drop(['label'], axis=1)
y = df['label']
```

```python
clf = DecisionTreeClassifier()
clf.fit(X, y)
data = export_graphviz(clf,
                       out_file=None,
                       feature_names=X.columns,
                       class_names=["TEL", "ALT"],
                       filled=True,
                       rounded=True,
                       special_characters=True)
graph = graphviz.Source(data)
graph
```

```python
rf = RandomForestClassifier(n_estimators=5000,oob_score=True, n_jobs=20)
rf.fit(X, y)
feature_importances = pd.DataFrame(
    rf.feature_importances_, index=X.columns,
    columns=['importance']).sort_values(
        'importance', ascending=False)
rf.oob_score_
```

```python
fig = plt.gcf()
fig.set_size_inches(6, 15)
ax = sns.barplot(x='importance', y='index', data=feature_importances.iloc[:30,:].reset_index())
```
