```python
import pandas as pd
from glob import glob
import re
pd.options.display.max_columns = 100
from tqdm import tqdm_notebook
```

```python
def map_name(name):
    name = name.replace("-", "_") 
    if name == "IIICF":
        name = "IIICF_P7"
    if name == "IIICFE6E7C4post":
        name = "IIICF_E6E7_C4_post"
    if name == "IIICFE6E7C4pre":
        name = "IIICF_E6E7_C4_pre"
    if name == "IVG_BF_LXSNpost":
        name = "IVG_BF_LXSN_post"
    if name == "IVG_BF_LXSNpre":
        name = "IVG_BF_LXSN_pre"
    if name == "JFCF_6_T_1_P":
        name = "JFCF_6_T_1_P_TEL"
    if name == "JFCF_6_T_1_PALT":
        name = "JFCF_6_T_1_P_ALT"
    if name == "JFCF6_P12_H06L4ALXX_6":
        name = "JFCF_6"
    if name in ("JFCF6_T_1J_1_3C_H06L4ALXX_3", "JFCF6_T_1J_1_3C"):
        name = "JFCF_6_T_1J_1_3C"
    if name in ("JFCF6_T_1J_6B_H06L4ALXX_4", "JFCF6_T_1J_6B"):
        name = "JFCF_6_T_1J_6B"
    if name == "JFCF_61M":
        name = "JFCF_6_T_1_M"
    if name == "JFCF_61Q":
        name = "JFCF_6_T_1_Q"
    if name == "LFS05F24post":
        name = "LFS_05F_24_post"
    if name == "LFS05F24pre":
        name = "LFS_05F_24_pre"
    if name == "MeT_4Apost":
        name = "MeT_4A_post"
    if name == "MeT_4Apre":
        name = "MeT_4A_pre"
    return name
```

```python
def split_gene(df, sep='&'):
    tmp = pd.DataFrame(columns=df.columns)
    for index, row in tqdm_notebook(df.iterrows()):
        genes = row['Gene']
        for gene in genes.split(sep):
            if gene == 'NONE':
                continue
            row['Gene'] = gene
            tmp = tmp.append(row)
    return tmp
```

```python
mutect = pd.read_csv(
    "/home/scai/processing_simoncai/tmm/tel_vs_alt/immortal_simon/annovar/combined_somatic.tsv",
    sep='\t')
mutect_filtered = mutect[~mutect['Func.refGene'].isin(['intronic','intergenic', 'ncRNA_intronic'])]

mutect_filtered = mutect_filtered.rename(
    columns={
        'sample': 'Name',
        'Gene.refGene': 'Gene',
        'Func.refGene': 'Type',
        'Chr': 'Chr1',
        'Start': 'Pos1',
        'Ref': 'Ref1',
        'Alt': 'Alt1'
    })

mutect_filtered['Source'] = 'MuTect2'
mutect_filtered['Key'] = mutect_filtered.apply(
    lambda row:
    f"MuTect2_{row['Type']}_{row['Gene']}_{row['Name']}_{row['Chr1']}_{row['Pos1']}_{row['Ref1']}_{row['Alt1']}",
    axis=1)

cols = ['Key', 'Name', 'Source']
cols = cols + [x for x in mutect_filtered.columns if 'Other' not in x and x not in cols]
cols.remove('End')

mutect_filtered = mutect_filtered[cols]

mutect_filtered['Name'] = mutect_filtered['Name'].map(lambda x:map_name(x))
```

```python
sv_df = pd.read_csv("../immortal_simon/gridss/DATA_GRIDSS.tsv", sep="\t").drop_duplicates()
```

```python
sv_df['lower'] = sv_df.apply(lambda row:min(int(row['Pos1'].split('-')[0]),int(row['Pos2'].split('-')[0])),axis=1)
sv_df['upper'] = sv_df.apply(lambda row:max(int(row['Pos1'].split('-')[1]),int(row['Pos2'].split('-')[1])),axis=1)


sv_df_same_chrom = sv_df[sv_df['Chr1']==sv_df['Chr2']]

sv_df_same_chrom = sv_df_same_chrom.reset_index(drop=True)

sv_df_bed = sv_df_same_chrom[['Chr1', 'lower', 'upper', 'Key']]

sv_df_bed['strand'] = '.'
```

```python
sv_df_bed.to_csv("../immortal_simon/gridss/sv_df.bed", index=False, header=False, sep='\t')
```

```python
sv_df_bed_exon = pd.read_csv(
    "../immortal_simon/gridss/sv_exon_coding_counts.bed",
    sep='\t',
    names=['chr', 'start', 'end', 'Key', 'strand', 'exon_count']).drop_duplicates(['Key'])
```

```python
sv_df_same_chrom_exon = pd.merge(sv_df_same_chrom, sv_df_bed_exon,
                                 on=['Key'
                                     ]).drop(['chr', 'start', 'end', 'strand', 'lower','upper'],
                                             axis=1)

sv_df_same_chrom_exon = sv_df_same_chrom_exon[sv_df_same_chrom_exon['exon_count'] > 0].reset_index(drop=True)

sv_df_exon = pd.concat([
    sv_df[sv_df['Chr1'] != sv_df['Chr2']].reset_index(drop=True),
    sv_df_same_chrom_exon.reset_index(drop=True)
],
                       sort=False).drop(['lower','upper'],axis=1)

sv_df_exon = sv_df_exon[sv_df_exon['Consequence']!='feature_fusion']
```

```python
sv_df_exon.to_csv("../immortal_simon/gridss/DATA_GRIDSS_Exon_coding.tsv", sep="\t", index=False)
```
