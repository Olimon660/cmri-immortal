```python
import glob
import pandas as pd
import numpy as np
```

```python
files = glob.glob("../data/CNA_results/segs/*")
```

```python
dfs = [pd.read_csv(f, sep='\t') for f in files]
```

```python
for df in dfs:
    to_del = []
    for i in range(1, df.shape[0]):
        if int(df.iloc[i, 2]) < int(df.iloc[i - 1, 3]) and df.iloc[i, 0] == df.iloc[
                i - 1, 0] and df.iloc[i, 1] == df.iloc[i - 1, 1]:
            to_del.append(i-1)
    df = df.drop(to_del)
```

```python
df_all = pd.concat(dfs)
```

```python
df_all['Sample'] = df_all['Sample'].str.replace('tumor_sample_', '')
```

```python
df_all.head()
```

```python
df_all['Sample'].unique()
```

```python
df_all['logR'] = df_all.apply(
    lambda row: np.log2(row['logR_Copy_Number'] / 2)
    if row['Chromosome'] != 'chrX' else np.log2(row['logR_Copy_Number'] / 1), axis=1)
```

```python
df_all = df_all[['Sample', 'Chromosome', 'Start_Position.bp.', 'End_Position.bp.', 'Length.snp.', 'logR']]
df_all['Chromosome'] = df_all['Chromosome'].str.replace('chr', '')
df_all['Chromosome'] = df_all['Chromosome'].str.replace('X', '23')
df_all['Chromosome'] = df_all['Chromosome'].str.replace('Y', '24')
```
